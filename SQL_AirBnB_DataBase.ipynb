```python
file_path = 'case_study_pmf.xlsx'  # Replace with the correct name
```

```python
import pandas as pd
```

```python
# Step 3: Load the Excel file
excel_data = pd.ExcelFile(file_path)

# Load each sheet into a DataFrame
listings_df = excel_data.parse('listings')
calendars_df = excel_data.parse('calendars')
reviews_df = excel_data.parse('reviews')
```

```python
import sqlite3

# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')
```

```python
# Save the parsed DataFrames as SQLite tables
listings_df.to_sql('listings', conn, index=False, if_exists='replace')
calendars_df.to_sql('calendars', conn, index=False, if_exists='replace')
reviews_df.to_sql('reviews', conn, index=False, if_exists='replace')
```

```python
import pandas as pd

query = "SELECT * FROM calendars LIMIT 50"
result = pd.read_sql_query(query, conn)

# Display the query result
print(result)
```

```python
#question 4(a)

query = """
SELECT
    l.property_type,
    AVG(c.price) AS average_price_per_night
FROM calendars c
JOIN listings l
ON c.listing_id = l.id
GROUP BY l.property_type
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, conn)

# Display the results
print(result)
```

```python
# Question - 4(b)
query = """
SELECT
    l.id AS listing_id,
    l.property_type,
    AVG(c.price) AS average_price_per_night
FROM calendars c
JOIN listings l
ON c.listing_id = l.id
GROUP BY l.property_type
ORDER BY average_price_per_night DESC
LIMIT 5;
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, conn)

# Display the results
print(result)
```
